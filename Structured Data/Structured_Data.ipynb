```python
Uid_grp = df.groupby("Patient-Uid")
target_index = []
for name,group in Uid_grp:
    indices = (group["Incident"] == "TARGET DRUG").idxmax()
    target_index.extend(group.loc[:indices].index)
```

```python
df_trgt = df.loc[target_index,:]
df_trgt
```

```python
features = pd.get_dummies(df_trgt.Incident).cumsum()
features
```

```python
target = df_trgt.Incident.map(lambda i : 1 if i =="TARGET DRUG" else 0)
target
```
